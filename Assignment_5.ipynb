```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("/content/Social_Network_Ads (1).csv")
df
```

```python
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df
```

```python
# Selecting features and target variable using column names
X = df[['Age', 'EstimatedSalary','Gender']].values  # Features: 'Age' and 'EstimatedSalary'
y = df['Purchased'].values                 # Target variable: 'Purchased'
```

```python
X
```

```python
y
```

```python
# Splitting the dataset into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#create instance of logistic regression
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(X_train,y_train)
```

```python
#predict labels for test set
y_pred = logistic_regression.predict(X_test)
```

```python
#create comfusion matrix
confusion =  confusion_matrix(y_test,y_pred)
```

```python
#display confusion matrix
print("Confusion Matrix:")
print(confusion)
```

```python
#extract values from confusion matrix
TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]
TP = confusion[1,1]
```

```python
#compute accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)
print("Accuracy is :", accuracy)
```

```python
#compute error rate
error_rate = (FP + FN) / (TP + TN + FP + FN)
print("Error rate is :", error_rate)
```

```python
#compute precision
precision = TP / (TP + FP) #positive predictions that are actually correct
print("Precision is :", precision)
```

```python
#display the recall
recall = TP / (TP + FN)   #actual  positive that are correctly identifies by model
print("Recall is :", recall)
```

FN (False Negative): These are the cases where the model incorrectly predicted the negative class. The model predicted Negative, but the actual value was Positive.

FP (False Positive): These are the cases where the model incorrectly predicted the positive class. The model predicted Positive, but the actual value was Negative.

TN (True Negative): These are the cases where the model correctly predicted the negative class. The model predicted Negative, and the actual value was also Negative.
