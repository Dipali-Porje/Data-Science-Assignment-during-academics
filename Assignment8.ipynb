```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean , inplace=True)
```

```python
df['Cabin'].dropna()
```

```python
df['Cabin'] = pd.to_numeric(df['Cabin'], errors='coerce')
```

```python

sns.countplot(x='Survived', data=df , color='lightblue')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Sex', data=df ,color='green')
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['Age'], bins=20, kde=True , color='orange')  # kde=True adds a kernel density estimate
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['Fare'] , color = 'pink')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=df, color = 'purple', alpha=0.6)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# 5. Count Plot for SibSp
sns.countplot(x='SibSp', data=df , color='r')
plt.title('Number of Siblings/Spouses aboard')
plt.xlabel('Siblings/Spouses Count')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming df is your DataFrame and it contains the 'Fare' column
sns.histplot(df['Fare'], bins=30, kde=True, color ='b')

# Add title and labels
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare ($)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
