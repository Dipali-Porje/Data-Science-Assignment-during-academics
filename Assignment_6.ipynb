```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB
```

```python
df= pd.read_csv("/content/Iris (1).csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
#split dataset
X= df.drop("Species", axis=1)
y= df["Species"]
```

```python
X
```

```python
y
```

```python
# Splitting the dataset into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initializing and training the Naïve Bayes classifier
naive_model = GaussianNB()
naive_model.fit(X_train, y_train)
```

```python
#predict labels for test set
y_pred = naive_model.predict(X_test)
```

```python
#create comfusion matrix
confusion =  confusion_matrix(y_test,y_pred)
```

```python
#display confusion matrix
print("Confusion Matrix:")
print(confusion)
```

Rows represent the true (actual) classes.

Columns represent the predicted classes.

Class 0: Setosa

Class 1: Versicolor

Class 2: Virginica


[[ 9  0  0]   <- True Setosa instances

 [ 0 10  1]   <- True Versicolor instances

 [ 0  0 10]]  <- True Virginica instances


 The diagonal elements represent correct predictions (True Positives).

The off-diagonal elements represent misclassifications (False Positives and False Negatives).

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Compute metrics
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred, average='macro')  # Using 'macro' for multi-class
recall = recall_score(y_test, y_pred, average='macro')        # Using 'macro' for multi-class
f1 = f1_score(y_test, y_pred, average='macro')                # Using 'macro' for multi-class

#macro average is a method of calculating performance metrics in multi-class classification problems, where you have more than two classes

# Print metrics
print(f"\n\nAccuracy is: {accuracy:.4f}")
print(f"\nError Rate: {error_rate:.4f}")
print(f"\nPrecision: {precision:.4f}")
print(f"\nRecall: {recall:.4f}")
print(f"\nF1 Score: {f1:.4f}")
```

alternate ways of evaluating metric

```python
#extract values from confusion matrix
TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]
TP = confusion[1,1]
```

```python
#compute accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)
print("Accuracy is :", accuracy)
```

```python
#compute precision
precision = TP / (TP + FP) #positive predictions that are actually correct
print("Precision is :", precision)
```

```python
#display the recall
recall = TP / (TP + FN)   #actual  positive that are correctly identifies by model
print("Recall is :", recall)
```
